```python
# ðŸš€ Step 1: Load Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸŽ¯ Step 2: Load the Dataset
df = pd.read_csv("train.csv")  # or change to your file name

# ðŸ“Š Step 3: Dataset Overview
print("ðŸ’¡ First 5 rows of the data:")
print(df.head())

print("\nðŸ“ˆ Shape of data:", df.shape)
print("ðŸ“‰ Total columns:", df.columns.tolist())

# ðŸ”Ž Step 4: Missing Data Check
print("\nðŸ§¹ Missing values in each column:")
print(df.isnull().sum())

# ðŸ› ï¸ Step 5: Data Cleaning (handling missing data)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column (too many missing values)
df.drop(columns='Cabin', inplace=True)

# âœ… Step 6: Cleaned Data Info
print("\nâœ… Data after cleaning:")
print(df.info())

# ðŸ“Š Step 7: Visualization - Survival Count
sns.set(style="whitegrid")
sns.countplot(x="Survived", data=df, palette="Set1")
plt.title("Passenger Survival Count")
plt.xticks([0,1], ['Not Survived','Survived'])
plt.show()

# ðŸŽ¯ Step 8: Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survival Rate by Gender")
plt.show()

# ðŸŽ¯ Step 9: Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set3')
plt.title("Survival Rate by Passenger Class")
plt.show()

# ðŸ§’ Step 10: Age Distribution
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.grid(True)
plt.show()

# ðŸ’¡ Step 11: Average Age by Survival
avg_age = df.groupby('Survived')['Age'].mean()
print("\nðŸ“Š Average Age of Passengers by Survival:")
print(avg_age)

# ðŸ“ Step 12: Final Notes
print("\nðŸ“Œ Insights:")
print("- Most survivors were female and in 1st/2nd class.")
print("- Many passengers were between 20-40 years old.")
print("- Older passengers were slightly less likely to survive.")
```
