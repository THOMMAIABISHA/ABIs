```python
# Step 1: Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Creating custom dataset (learned from YouTube )
data = {
    'Age_Group': ['0-14', '15-24', '25-54', '55-64', '65+'],
    'Male_Population': [15, 10, 25, 7, 5],
    'Female_Population': [14, 9, 24, 8, 6]
}

df = pd.DataFrame(data)

# Step 3: Plotting grouped bar chart
x = np.arange(len(df['Age_Group']))  # label locations
width = 0.35  # width of the bars

# Step 4: Creating the bar chart
plt.figure(figsize=(9,6))
plt.bar(x - width/2, df['Male_Population'], width, label='Male', color='royalblue')
plt.bar(x + width/2, df['Female_Population'], width, label='Female', color='lightcoral')

# Step 5: Adding labels and title
plt.xlabel('Age Group')
plt.ylabel('Population (in Millions)')
plt.title('Gender-wise Population Distribution by Age Group')
plt.xticks(x, df['Age_Group'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Step 6: Showing the plot
plt.show()
```
